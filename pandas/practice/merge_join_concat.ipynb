```python
import pandas as pd
```

### Concatinating objects

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])
```

### DataFrame concat
pandas.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=None, copy=True)

```python
# concat df1, df2, df3 with axis = 0
frames = [df1, df2, df3]
result = pd.concat(frames)
result
```

```python
# associate a specific key to the sub dfs in the result df
result = pd.concat(frames, keys=['x', 'y', 'z'])
result
```

```python
# accessing sub subframe with a key
result.loc['y']
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
```

```python
result = pd.concat([df1, df4], axis = 1)
result
```

```python
result = pd.concat([df1, df4], axis = 1, join = 'inner')
result
```

```python
result = pd.concat([df1, df4], axis = 1, join = 'outer')
result
```

```python
result = pd.concat([df1, df4], axis = 1, join_axes = [df1.index])
result
```

```python
result = pd.concat([df1, df4], axis=0, ignore_index=True, sort=False)
result
```

```python
# concat series with a dataframe
s1 = pd.Series(['x1', 'x2', 'x3'], name='S')
result = pd.concat([df1, s1], axis=1)
result
```

```python
pieces = {'x': df1, 'y': df2, 'z': df3}
result = pd.concat(pieces, axis=0)
result
```

```python
result = pd.concat(pieces, axis=0, keys=['y', 'z'])
result
```

### database style DataFrame joining/merging
pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None,
         left_index=False, right_index=False, sort=True,
         suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
result = pd.merge(left, right, how='inner', on='key')
result
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
result = pd.merge(left, right, how='inner', on=['key1', 'key2'])
#result = pd.merge(left, right, how='outer', on=['key1', 'key2'])
#result = pd.merge(left, right, how='left', on=['key1', 'key2'])
#result = pd.merge(left, right, how='right', on=['key1', 'key2'])
result
```

```python
left  = pd.DataFrame({'A' : [1,2],   'B' : [2,2]})
right = pd.DataFrame({'A' : [4,5,6], 'B' : [2,2,2]})

# duplicate join keys in DataFrames
result = pd.merge(left, right, on='B', how='outer')
result
```

```python
# checking for duplicate keys
left  = pd.DataFrame({'A' : [1,2],   'B' : [1,2]})
right = pd.DataFrame({'A' : [4,5,6], 'B' : [2,2,2]})

#result = pd.merge(left, right, on='B', how='outer')
result = pd.merge(left, right, on='B', how='outer', validate="one_to_many")
result
```

### join

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'key': ['K0', 'K1', 'K0', 'K1']})

right = pd.DataFrame({'C': ['C0', 'C1'],
                      'D': ['D0', 'D1']},
                      index=['K0', 'K1'])

result = left.join(right, on='key', how='outer')
result
```
